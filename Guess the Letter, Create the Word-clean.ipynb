**Create a Letter Guessing Game**

Create a program that allows a user 7 chances to guess the correct word. If they do not guess the word after 7 tries, the player loses and the program will print the correct word. Otherwise, the play wins and the game.

**How to build the program**
* When you start the game, you will need to select a random word from a list of at least 10 words (You have full control over which words you want to use for you program). This will be your secret word. Your secret word will be represented in the program as a group of underscores. For as long as the word is, you should also have that many underscores. 

* Once the word is selected, your game will commence. Perform a Google search to figure out how to select a random word from a list using Python.<br>
**Hint** There's package you can import into your application that does this for you.

* The end user will have a total of 7 chances to guess the correct letter from the secret word. If the end user makes 7 incorrect guesses, the game will end.

* As you guess the correct letters, the letters you have guess will then take place of the underscores that letter represents. <br>
**For Example**: If your secret word is 'watermelon' and  so far you have guessed the letters 'a' and 'e', the word you're trying to guess will appear as follows: _ a _ e _ _ e _ _ _.<br>
**Keep in mind** that if you guess a letter that appears more than once in your secret word, make sure that the letter is populated anywhere that letter would be.

```python
# 1. Make a program that selects a word from a list of words

# 2. From each category, select a random word

# 3. Accept user input that asks the user to type in a category

# randomly_selected_word = "orange"

# 4. _ _ _ _ _ _

# 5. if you make a correct choice = "o" => o _ _ _ _ _


# randomly_selected_word = "watermelon"

# 6. _ _ _ _ _ _ _ _ _ _ _

# 7. if you make a correct choice = "e" => _ _ _ e _ _ e _ _ _

# 8. 7 attempts

# 9. if you exhausted all 7 choices, show the correct word 
```

```python
import random_generator
from word_library import word_list

def Random_Word():
    word = random_generator.choice(word_list)
    # ^this random.choice function will pick a word from the list at random.
    return word.upper()
    # ^this will return the word thats selected in uppcase letters for the user to see.

def Game_Ruels(word): # in this defined function, we are referencing the word that was selected.
    word_selction = "_" * len(word) # This will be the placeholder for the word, replacing the lettes of the word by "_ _ _ _ _ _ _ _ *underscores*"
    guesses = False
    guess_letters = [] # This empty list will store guessed letters by the user
    guess_words = [] # This empty list will store guessed words by the user
    attempts = 7 # This is how many attempts the user has left until they lose the game
    print("Hangman")
    print(f"You have {attempts} attempts left...") # Show how many attempts the user has left
    print(word_selction) # This will display the word in "_ _ _ _ _ " and any letters they have gotten right
    print("\n")

    while not guesses and attempts > 0: # This means while the user has attempts left not equaling zero, they can continue to play
        guess = input("Guess a letter or a word:\n").upper()

        if len(guess) == 1 and guess.isalpha(): # This means, if "1" character is typed, this is a letter guess. *isalpha means it has a to a letter*
            if guess in guess_letters: # basically, if guessed letter has already been used, try again.
                print("This letter has already been used...", guess)
            elif guess not in word: # if guess wasn't in the word, subtract the amount of attempts the user has left and append the guess to guess_letters list.
                print(f"{guess} is not in the word.")
                attempts -= 1
                guess_letters.append(guess)
            else: # This means that the user got the letter correct and gets to continue with same amount of attempts
                print(f"Correct, {guess} is in the word!")
                guess_letters.append(guess) # This will append the letter to the empty letter list
                guess_word_list = list(word_selction)
                indices = [i for i, letter in enumerate(word) if letter == guess]
                # Indices are the indexes that are being referenced
                # Enumerate() method adds a counter to an iterable and returns it in a form of enumerating object. This enumerated object can then be used directly for loops or converted into a list of tuples using the list() method.
                for index in indices:
                    guess_word_list[index] = guess # This will show the letter guessed from the guess input.
                word_selction = "".join(guess_word_list) # This will append the letter to the empty letter list and convert the enumerate object back to a string
                if "_" not in word_selction: 
                    guess = True # This means the letter will replace the "_" character with the letter

        elif len(guess) == len(word) and guess.isalpha(): # This means, if more than "1" character is typed, user guessed a word.
            if guess in guess_words:
                print(f"{guess} has already been guessed!") # The user has already guessed this exact word
            elif guess != word: # This means the user answered with a wrong word.
                print(f"{guess} is not the word")
                attempts -= 1 # This means the user will lose an attempt
                guess_words.append(guess) # This will append the word to the empty word list
            else:
                guesses = True # This means the user guessed the word successfully
                word_selction = word # This will determine that the game was won by the user

        else: # if the user makes it to this stage, the guess was neither a letter or word.
            print("This is not a valid guess, try again!")
            print(f"You have {attempts} attempts left...")
            print(word_selction)
            print("\n")

    if guesses:
        print("Congragulatios, you guessed the word correctly! You win the game!") # if player guessed letters or word correctly, they won.
    else: 
        print(f"Sorry, you ran out of attempts. The word was {word}.") # If the user didn't guess the word correctly, the game was lost and the word will be shown.

def Play_Game(): # This function is the driver for the game, meaning it will run the functions.
    word = Random_Word()
    Game_Ruels(word)
    while input("Would you like to play again? (Y)es | (N)o").upper() == "Y":
        word = Random_Word()
        Game_Ruels(word)

if __name__ == "__main__":
    Play_Game()
    
```
